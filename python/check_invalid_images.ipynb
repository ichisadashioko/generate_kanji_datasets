```python
img_dir = '/media/shioko/SHIOKO/kanji_images'
```

```python
def check_invalid_image(path):
    if os.path.isdir(path):
        file_list = os.listdir(path)
        for file in file_list:
            file_path = os.path.join(path, file)
            check_invalid_image(file_path)
            
    else:
        queue_img = Image.open(path)
        diff = ImageChops.difference(img, queue_img)
        np_diff = np.array(diff)
        if np.sum(np_diff) < 255:
            print(path)
            os.remove(path)
```

```python
check_invalid_image(img_dir)
```

```python
def remove_invalid_image(path):
    if os.path.isdir(path):
        file_list = os.listdir(path)
        for file in file_list:
            file_path = os.path.join(path, file)
            remove_invalid_image(file_path)
            
    else:
        if 'ProN_R_ttf' in path:
#             print(path)
            os.remove(path)
```

```python
remove_invalid_image(img_dir)
```
